```python
def get_eigen_values_vectors(matrix):
    eigenvalues, eigenvectors = LA.eig(matrix)
    indexes = []
    
    for i in range(eigenvalues.shape[0]):
        l = eigenvalues[i]
        v = eigenvectors[:, i]
        A_v = matrix @ v
        l_v = l * v
        
        if not np.allclose(A_v, l_v):
            pass
        else:
            indexes.append(i)
    
    return eigenvalues[indexes], eigenvectors[:, [indexes]]
            
```

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from matplotlib.image import imread
from skimage import color
```

```python
img = imread('ukhiliant-chomu-trend-na-ukhilennia-vid-sluzhbi-nabiraie-populiarnosti-v_1.webp')
```

```python
plt.imshow(img)
plt.axis('off')
plt.show()
img.shape
```

```python
image_sum = img.sum(axis=2)
image_bw = image_sum / image_sum.max()

plt.imshow(image_bw, cmap='gray')
plt.axis('off')
plt.show()
image_bw.shape
```

```python
image_bw_flat = image_bw.flatten()
```

```python
pca = PCA(random_state=42)
```

```python
pca.fit(image_bw_flat.reshape(-1, 1))

cumulative_variance = np.cumsum(pca.explained_variance_ratio_) * 100

num_components = np.argmax(cumulative_variance >= 95) + 1

print(f'Кількість компонентів для покриття 95% дисперсії: {num_components}')

plt.figure(figsize=(10, 6))
plt.plot(cumulative_variance, marker='o', linestyle='--')
plt.axhline(color='r', linestyle='-')
plt.axvline(x=num_components, color='k', linestyle='--')
plt.xlabel('Головні компоненти')
plt.ylabel('Кумулятивна дисперсія (%)')
plt.title('Кумулятивна дисперсія, пояснена головними компонентами')
plt.grid()
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(image_bw_flat)
image_scaled = scaler.transform(image_bw_flat)
```

```python
pca = PCA(n_components=.95, random_state=42)
image_bw_pca = pca.fit_transform(image_bw)

image_bw_reconstructed = pca.inverse_transform(image_bw_pca)


plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.title('Оригінальне зображення')
plt.axis('off')
plt.imshow(image_bw, cmap='gray')

plt.subplot(1, 2, 2)
plt.title('Відновлене зображення після PCA')
plt.imshow(image_bw_reconstructed, cmap='gray')
plt.axis('off')
plt.show()
```

```python
pca.n_components_
```

```python
def plot_reconstructed_image(img, n_components):
    plt.figure(figsize=(6, 6))
    plt.title(f'Count of components: {n_components}')
    plt.axis('off')
    plt.imshow(img, cmap='gray')
    plt.show()
```

```python
for n in range(1, 110, 10):
    pca = PCA(n_components=n, random_state=42)
    image_bw_pca = pca.fit_transform(image_bw)
    image_bw_reconstructed = pca.inverse_transform(image_bw_pca)
    
    plot_reconstructed_image(image_bw_reconstructed, n)
```

```python
pca = PCA(n_components=100, random_state=42)
image_bw_pca = pca.fit_transform(image_bw)
image_bw_reconstructed = pca.inverse_transform(image_bw_pca)
```

```python
def encrypt_message(message, key_matrix):
    message_vector = np.array([ord(char) for char in message])
    eigenvalues, eigenvectors = np.linalg.eig(key_matrix)
    diagonalized_key_matrix = np.dot(np.dot(eigenvectors, np.diag(eigenvalues)), np.linalg.inv(eigenvectors))
    encrypted_vector = np.dot(diagonalized_key_matrix, message_vector)
    return encrypted_vector
```

```python
message = 'Hello, World!'
```

```python
key_matrix = np.random.randint(0, 256, (len(message), len(message)))
```

```python
encrypted = encrypt_message(message, key_matrix)
encrypted
```

```python
def decrypt_message(encrypted_vector, key_matrix):
    eigenvalues, eigenvectors = np.linalg.eig(key_matrix)
    diagonalized_key_matrix = np.dot(np.dot(eigenvectors, np.diag(eigenvalues)), np.linalg.inv(eigenvectors))
    decrypted_vector = np.dot(np.linalg.inv(diagonalized_key_matrix), encrypted_vector)
    decrypted_message = ''.join([chr(int(round(char))) for char in decrypted_vector])
    return decrypted_message
```
